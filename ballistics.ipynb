# Ballistics and the Range Equation
This is an investigation of the so-called *range equation*: $d=\frac {v^2} g \sin 2\theta$.

```python
from math import radians, sin, pi
from numpy import linspace
import matplotlib.pyplot as plt
```

```python
def rangeformula(v, theta):
    d = (((v * v)/9.81)*sin(2*theta))
    return d
    """
    Inputs:
    v - velocity in m/s
    theta - launch angle with respect to horizon in radians
    Returns:
    Impact distance in m
    """
```

```python
angledegrees = range(0,91)
```

```python
rangefromdegrees = [rangeformula(100, radians(x)) for x in angledegrees] 
```

```python
plt.plot(angledegrees, rangefromdegrees)
plt.title('Range as a function of angle at 100 m/s')
plt.xlabel('Angle (degrees)')
plt.ylabel('Range (meters)')
plt.show()
```

```python
angledegrees = linspace(0, 90, num=200)
```

```python
rangeforangle100 = [rangeformula(100, radians(angle)) for angle in angledegrees]
rangeforangle200 = [rangeformula(200, radians(angle)) for angle in angledegrees]
plt.plot(angledegrees, rangeforangle100, label="100 m/s")
plt.legend()
plt.plot(angledegrees, rangeforangle200, label="200 m/s")
plt.legend()
plt.title('Range as a function of angle at various velocities')
plt.xlabel('Angle (degrees)')
plt.ylabel('Range (meters)')
plt.show
```

```python
def rangeformula(v, theta):
    d = (((v * v)/9.81)*sin(2*radians(theta)))
    return d
    """
    Inputs:
    v - velocity in m/s
    theta - launch angle with respect to horizon in radians
    Returns:
    Impact distance in m
    """
```

```python
velocityrange = linspace(0, 100, num=200)
```

```python
rangefromvelocity = [rangeformula(x, 45) for x in velocityrange]
```

```python
plt.plot(velocityrange, rangefromvelocity)
plt.show()
```

```python
plt.plot(velocityrange, ([rangeformula(x, 15) for x in velocityrange]))
plt.plot(velocityrange, ([rangeformula(x, 30) for x in velocityrange]))
plt.plot(velocityrange, ([rangeformula(x, 45) for x in velocityrange]))
plt.plot(velocityrange, ([rangeformula(x, 60) for x in velocityrange]))
plt.plot(velocityrange, ([rangeformula(x, 75) for x in velocityrange]))
plt.show()
```
