# Interest Table Notebook:
This notebook was made in order to generate a table and accompanying graph for this interest equation, at which the money invested starts at 1000 dollars and increases at 3.5% yearly:

$i=1000(1 + \frac {3.5} {100})^t$

```python
from numpy import linspace
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame()
df['years'] = linspace(0, 30, 50)
df['value_float'] = 1000*(1.035)**df.years
df['value_str'] = "$" + str(round(df.value_float, 0))
```

```python
plt.scatter(df.years, df.value_float)
plt.xlabel('years')
plt.ylabel('dollars')
plt.show
```
